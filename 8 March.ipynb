## Check Normal Distribution using QQ-Plot Quartile - Quartile plot

```python
import scipy.stats as stat
import pylab
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
#We will check for every features whether it belongs to normal distribution or not
import matplotlib.pyplot as plt
```

```python
def plot_data(df,feature):   #using this function we will try to make some figure
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)    # no of plots
    # Create 2 figure one on left and one on right
    #left will show histogram or distribution and the right will show thw QQ plot
    # df[feature].hist() #Construct histogram
    sns.histplot(df[feature],kde=True)   #It   will looks like an normal distribution
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)  # dist will compare x variable with other variable
    plt.show() 
```

```python
#How figure will come
plot_data(df,'sepal_width')
```

```python
#We can observe if the right side diagram will come in a straight line then it will be a normal distribution
#In Q Q plot x axis has normal distribution and y axis has the feature value 
```

```python
plot_data(df,'petal_width')
```

```python
#This is not a normal or gaussian distribution (points are not on a straight line)
```

```python
plot_data(df,'sepal_length')
```

```python
#This also follows the normal distribution
```

```python
plot_data(df,'petal_length')
```

```python
#This is not a normal or gaussian distribution
```
